```python
#Importing Seaborn & Matplotlib for DataViz
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Histograms for the different variables
#Defining elements of the grid
sns.set(style="darkgrid")
f, ax = plt.subplots(2, 3, figsize=(15, 9))

#Defining the palette
palette = ["#F72585", "#7209B7", "#3A0CA3", "#4361EE", "#4CC9F0", '#33658A']

sns.histplot(data = df, x="age", label = 'Age', color = palette[0], ax = ax[0,0])
sns.histplot(data = df, x="orders", label = 'Orders', color = palette[1], ax = ax[0,1])
sns.histplot(data = df, x="discount_rate", label = 'Discount rate',  color = palette[2], ax = ax[0,2])
sns.histplot(data = df, x="revenue", label = 'Revenue', color = palette[3], ax = ax[1,0])
sns.histplot(data = df, x="average_item_value", label = 'Average item value', color = palette[4], ax = ax[1,1])
sns.histplot(data = df, x="profit",label = 'Profit', color = palette[5], ax = ax[1,2])

plt.show()
```

```python
df = df[df['age'] < 90]
```

```python
df = df[df['revenue'] < 500]
```

```python
df.describe()
```

```python
aggregated_rev = []
for j in df.columns[7:-2]:
    aggregated_rev.append([j, df[j].sum()])
```

```python
print(aggregated_rev)
labels = [element[0] for element in aggregated_rev]
values = [element[1] for element in aggregated_rev]
```

```python
len(labels)
```

```python
fig, ax = plt.subplots()
ax.pie(values, labels=labels, autopct='%1.1f%%')
```

```python
fig, ax = plt.subplots()
bar_colors = ['#EAF2F8', '#D4E6F1', '#A9CCE3', '#7FB3D5', '#5499C7', '#2980B9', '#2874A6', '#2471A3', '#1F618D', '#1A5276', '#154360']
fig.set_size_inches(20, 6)
ax.bar(labels, values, label=labels, color = bar_colors, align = 'center', width = 0.7)

plt.show()
```
